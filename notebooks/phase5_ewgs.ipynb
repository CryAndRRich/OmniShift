# Phase 5 — EWGS Gradient Estimator

**Base**: Phase 4 (SparseShift + PoT-BN)
**Change**: Replace STE backward with EWGS in all quantizers

EWGS formula: `g = g_STE ⊙ (1 + λ · sign(g_STE) ⊙ sign(w − Q(w)))`, λ=0.02

The forward pass is **byte-for-byte identical** to Phase 4 — inference energy unchanged.
Improvement shows as higher accuracy from smoother gradients near quantization boundaries.

Models:
- `sparseshift_fixed50_potbn_w30_ewgs` — fixed 50% sparsity
- `sparseshift_learnable_potbn_w30_ewgs` — learnable sparsity
